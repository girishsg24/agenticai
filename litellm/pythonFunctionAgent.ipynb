````python
from litellm import completion
from typing import List, Dict


def extract_code_block(response_text):
    match = re.search(r"```python(.*?)```", response_text, re.DOTALL)
    if match:
        return match.group(1).strip()
    return None


def generate_response(messages: List[Dict]) -> str:
    """Call LLM to get response"""
    response = completion(
        model="openai/gpt-4o",
        messages=messages,
        max_tokens=1024
    )
    return response.choices[0].message.content


what_func_to_create = input("What function do you want to create")

messages = [
    {"role": "system", "content": "You are a helpful agent who generates code in python"},
    {"role": "user", "content": "write a function for the following description " + what_func_to_create}
]
function_response = generate_response(messages)
parsed_function = extract_code_block(function_response)
messages = [
   {"role": "system", "content": "You are a helpful agent who generates code in python"},
   {"role": "assistant", "content": parsed_function},
   {"role": "user", "content": "add comprehensive documentation for the generated function including Function description" +
    "Parameter descriptions, Return value description, Example usage, Edge cases"
    }
]
doc_response = generate_response(messages)
messages.append({"role": "assistant", "content": doc_response})
messages.append({"role": "user", "content": "create unit test cases using the python unit test framework covering basic functionality,"+
                 " edge cases, error cases & various input scenarios"})
unit_test_response = generate_response(messages)
print(parsed_function)
print(doc_response)
print(unit_test_response)
````
